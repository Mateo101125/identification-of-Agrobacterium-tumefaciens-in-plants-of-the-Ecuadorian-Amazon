```python
### Bixa orellana chromosome Chloroplast DNA
from Bio import Entrez #importo el Entrez
from Bio import SeqIO #importo el SeqIO

Entrez.email = "A.N.Other@example.com"
handle=Entrez.efetch (db="nucleotide", rettype="fasta", retmode="text", id="MH025909.1")
record = SeqIO.read(handle, "fasta")
handle.close()
print(record.description)
print(record.seq)
bixa_sequence= record.seq
```
